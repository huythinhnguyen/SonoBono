```python
import os
from matplotlib import pyplot as plt

path_parent = os.path.dirname(os.getcwd())
os.chdir(path_parent)
%matplotlib inline

import numpy as np

from echo_lib.echo_transform_utils import *
```

Get the reference echo of pole at 1 meter distance, 0 degree, raw

```python
pole1m_raw = get_raw_echo(1, 1.0, 0.0, random_mode = False, index=0)
```

```python
%matplotlib notebook
```

```python
plt.rcParams.update({'font.size': 26})
#plt.rcParams.update({'font.weight': 'normal'})
fig, ax = plt.subplots(figsize = (6,5))
ax.plot(d_axis, pole1m_raw['left'], linewidth = 0.5, alpha=0.5)
ax.plot(d_axis, pole1m_raw['right'], linewidth = 0.5, alpha=0.5)
ax.set_xlim(0,3.5)
ax.set_ylim(-1000,1000)
ax.legend(['left', 'right'])
ax.set_xlabel('Distance (meter)')
ax.set_ylabel('Raw waveform (a.u.)')
ax.ticklabel_format(axis='y',style='sci', scilimits=(0,1))
fig.tight_layout()
```

```python
plant1m_raw = get_raw_echo(2, 1.0, 0.0, random_mode = False, index=0)
```

```python
plt.rcParams.update({'font.size': 26})
#plt.rcParams.update({'font.weight': 'normal'})
fig, ax = plt.subplots(figsize = (6,5))
ax.plot(d_axis, plant1m_raw['left'], linewidth = 0.5, alpha=0.5)
ax.plot(d_axis, plant1m_raw['right'], linewidth = 0.5, alpha=0.5)
ax.set_xlim(0,3.5)
ax.set_ylim(-1000,1000)
ax.legend(['left', 'right'])
ax.set_xlabel('Distance (meter)')
ax.set_ylabel('Raw waveform (a.u.)')
ax.ticklabel_format(axis='y',style='sci', scilimits=(0,1))
fig.tight_layout()
```

```python
pole1m_envelope = echo2envelope(pole1m_raw)
```

```python
fig, ax = plt.subplots(figsize = (6,5))
ax.plot(d_axis, pole1m_envelope['left'], linewidth = 2, alpha=0.8)
ax.plot(d_axis, pole1m_envelope['right'], linewidth = 2, alpha=0.8)
ax.set_xlim(0,3.499)
ax.set_ylim(0,6)
ax.legend(['left', 'right'])
ax.set_xlabel('Distance (meter)')
ax.set_ylabel('Envelope (a.u.)')
ax.ticklabel_format(axis='y',style='sci', scilimits=(0,1))
fig.tight_layout()
```

```python
plant1m_envelope = echo2envelope(plant1m_raw)
```

```python
fig, ax = plt.subplots(figsize = (6,5))
ax.plot(d_axis, plant1m_envelope['left'], linewidth = 2, alpha=0.8)
ax.plot(d_axis, plant1m_envelope['right'], linewidth = 2, alpha=0.8)
ax.set_xlim(0,3.499)
ax.set_ylim(0,6)
ax.legend(['left', 'right'])
ax.set_xlabel('Distance (meter)')
ax.set_ylabel('Envelope (a.u.)')
ax.ticklabel_format(axis='y',style='sci', scilimits=(0,1))
fig.tight_layout()
```

```python
pole1m_z = echo2envelope_z(pole1m_raw)
plant1m_z = echo2envelope_z(plant1m_raw)
```
